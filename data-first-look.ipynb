```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
# skiprows=600 ignores the first 600 lines
# sep='\t' assumes a tab separator; change to ',' if it's a true CSV
df = pd.read_csv(
    'data-mar12.csv', 
    skiprows=600, 
    sep=',', 
    header=None, 
    names=['timestamp', 'value']
)

# Convert the timestamp column to actual datetime objects
df['timestamp'] = pd.to_datetime(df['timestamp'])

print(df.head())
```

```python
# 3. Create the plot
plt.figure(figsize=(12, 6))
plt.scatter(df['timestamp'], df['value'])

# 4. Formatting the chart
plt.title('Value vs Time')
plt.xlabel('Timestamp')
plt.ylabel('Value (mm)')
plt.grid(True, linestyle='--', alpha=0.7)

# Rotate date labels for better readability
plt.xticks(rotation=45)
plt.tight_layout()

# 5. Display the plot
plt.show()
```

```python
df2 = pd.read_csv(
    'data-mar16.csv', 
    skiprows=1, 
    sep=',', 
    header=None, 
    names=['timestamp', 'P1', 'P2', 'P3', 'P4', 'P5', 'P6', 'P7', 'P8', 'P9', 'P10', 'P11', 'P12', 'P13' ]
)
```

```python
for col in df2.columns[1:]:
    plt.scatter(df2["timestamp"], df2[col], marker='o', label=col)

plt.xlabel("Time")
plt.ylabel("Pressure Value")
plt.title("Sensor Pressure Readings")
plt.legend(ncol=3)
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```
